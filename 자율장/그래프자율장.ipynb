```python
import math

parent = dict()

def get_dist(p1,p2):
    a = p1[0] - p2[0]
    b = p1[1] - p2[1]
    return math.sqrt((a*a)+(b*b))

def get_parent(parent,n):
    if parent[n] != n:
        parent[n] = get_parent(parent,parent[n])
    return parent[n]

def union(parent,a,b):
    a = get_parent(parent,a)
    b = get_parent(parent,b)
    
    if a < b:
        parent[a] = b
    else:
        parent[b] = a
        
def find(parent,a,b):
    a = get_parent(parent,a)
    b = get_parent(parent,b)
    
    if a == b:
        return True
    else:
        return False
```
