### Counter구조
#### Counter는 dict의 subclass로써 새로운 인스턴스를 만드는 클래스

### Counter 생성예시
#### Counter 클래스로 생성하는 이유는 실제 키값들에 연속된 상황이 확인이 필요한 경우 사용

```python
import collections

print(collections.Counter("attacked"))

print(collections.Counter({1:2,2:2}))

print(collections.Counter({1:2,2:2}.items()))

print(collections.Counter([1,2,3]))
```

### Counter 사칙연산
#### Counter 인스턴스 내의 키값이 같은 경우에 +/- 연산이 가능하며 zero값은 표시하지 않음

```python
from collections import Counter

a = Counter("abceabde")
b = Counter("abcdef")

# Counter 더하기
add = a + b
print("a + b : ", add)

# Counter 빼기
sub1 = a - b
print("a - b :", sub1)

sub2 = b - a
print("b- a:", sub2)

```

#### Counter 인스턴스 내의 키값이 같은 경우에 & / | 연산이 가능

```python
from collections import Counter

s = Counter("abceabde")
s2 = Counter("abcdef")

print("s : ",s)
print("s2 : ",s2)

# 교집합
inter = s & s2
print("s & s2 : ",inter)

# 합집합
union = s | s2
print("s | s2 : ",union)
```

```python
for i in union.elements():
    print(i)
```
