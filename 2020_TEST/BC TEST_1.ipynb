### 내 풀이

```python
n = ["가을","우주","너굴"]
n1 = ["너굴","너울"]
n2 = ["가봄","여울","봄"]
```

```python
def solution(name_list):
    name_list.sort()
    for i in range(len(name_list)-1):
        for j in range(i+1,len(name_list)):
            if name_list[i] in name_list[j] or name_list[j] in name_list[i]:
                if name_list[i] != "" and name_list[j] =="":
                    return False
                elif name_list[j]=="" and name_list[i]=="":
                    return False
                return True
    return False
```

```python
solution(n)
```
