```python
def solution(p_book):
    for i in range(len(p_book)-1):
        for j in range(i+1,len(p_book)):
            if p_book[i] in  p_book[j][:len(p_book[i])] or p_book[j] in p_book[i][:len(p_book[j])]:
                return False
    return True
```

#### 정답 2
    > 전형적인 hashMap 사용해서 풀이
    > 다른사람의 풀이

```python
def solution(phone_book):
    answer = True
    hash_map = {}
    for phone_number in phone_book:
        hash_map[phone_number] = 1
    for phone_number in phone_book:
        temp = ""
        for number in phone_number:
            temp += number
            if temp in hash_map and temp != phone_number:
                answer = False
    return answer
```

#### 정답3
    > B.startswith(A) = B문자열이 A문자열로 시작하는지 확인하는 메소드
    

```python
def solution(phoneBook):
    phoneBook = sorted(phoneBook)

    for p1, p2 in zip(phoneBook, phoneBook[1:]):
        if p2.startswith(p1):
            return False
    return True
```

```python
from collections import Counter

def solution(clothes):
    counter_each_category = Counter([cat for _, cat in clothes])
    all_possible = 1
    print(counter_each_category)
    for key in counter_each_category:
        all_possible *= (counter_each_category[key] + 1)
        
    return all_possible-1
```
